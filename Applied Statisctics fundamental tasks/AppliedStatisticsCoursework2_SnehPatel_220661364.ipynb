```python
#Using cdf method of scipy.stats.beta made a function name beta_measure which takes two arguments a, b and returns the probability mass
def beta_measure(a,b):
    x = np.linspace(a,b,100)
    beta = stats.beta.cdf(x,3,7)
    print(beta)
    plt.plot(beta, label="Beta(3,7)")
    plt.title("Beta Distribution with beta_measure(a,b)")
    plt.legend()
```

```python
#defining a pdf function which takes argument x and returns the pdf of the probability measure Beta (3, 7) evaluated at x 
def beta_pdf(x):
    return stats.beta.pdf(x, 3, 7, loc=0, scale=1)
```

```python
#By using quad we can derive the pdf function at the given intervals
result = integrate.quad(lambda x: beta_pdf(x), 0, 1)
print("For 0,1 the result is", result)
```

```python
result = integrate.quad(lambda x: beta_pdf(x), 0, 0)
print("For 0,0 the result is", result)
```

```python
result = integrate.quad(lambda x: beta_pdf(x), 0.25, 0.75)
print("For 0.25,0.75 the result is", result)
```

```python
result = integrate.quad(lambda x: beta_pdf(x), 0, 0.5)
print("For 0,0.5 the result is", result)
```

```python
result = integrate.quad(lambda x: beta_pdf(x), 0.5, 1)
print("For 0.5,1 the result is", result)

#From the results of pdf function derivation at the different intervals we can get quiet similar values that match with the probability masses of the cdf function which we calculated above.From this we can say that probability mass is nothing but area under the curve within the range [a,b] and in the given same interval the values will also tend to match
```

```python
#if a probability distribution d1 has density f1 and a probability distribution & d2 has density f2, then the density of the sum d1 +d2 is given by the convolution of the two densities
#Here we consider Beta(3,7) & Beta(7,3) as the substitute function so the sum of those will be same formula for d1 & d2

print("The support of a beta distribution is it's value in the range -> [0,1]")

#Calculating the support of Beta(3, 7)

cdf1 = stats.beta.cdf(0, 3, 7)
cdf2 = stats.beta.cdf(1, 3, 7)
support_at_37 = cdf2 - cdf1
print("Support of Beta(3, 7) is: ", support_at_37)


#Calculating the support of Beta(7, 3)

cdf1 = stats.beta.cdf(0, 7, 3)
cdf2 = stats.beta.cdf(1, 7, 3)
support_at_73 = cdf2 - cdf1
print("Support of Beta(7, 3) is: ", support_at_73)
```

```python
#Defining the integrand of the integral for the support of sum of both functions 
def integrationn(x, t):  
  y = stats.beta.pdf((t - x), 7, 3)
  z = stats.beta.pdf(x, 3, 7)
  return z*y

#integrating over x
integrand = quad(integrationn, 0, 1, args = (1))  
print("Support of Beta(3, 7) + Beta(7, 3) is: ", integrand[0])
```

```python
#compute the pdf f1+2(ti) at these 100 points using quad
t = np.linspace(0, 2, 100)
x1 = []
for i in t:
  z1 = quad(integrationn, 0, 1,args = (i))
  x1 = np.append(x1, z1[0])
    
#Pdf for Beta (3, 7) plus Beta (7, 3) at the 100 points
plt.plot(t, x1) 
```

```python
# Generating 1000 samples for the same

f1 = np.random.beta(3, 7, size = 10000)
f2 = np.random.beta(7, 3, size = 10000)
s1 = f1 + f2
print(s1)
sb.distplot(s1)
plt.show()


#It is clear from the graph above that, due to the large number of sampling done on it, the histogram is smoothened and it approximately gives us the curve of the PDF which was calculated in the previous step.
#Hence, the PDF of Beta(3, 7) plus(+) PDF of Beta(7, 3) = addition of random sampling from both our distributions.
```

```python
#get the data and load it
my_data = pdr.DataReader('MSFT', 'yahoo', '2012-11-02', '2022-11-02')
```

```python
my_data
```

```python
#keeping the close column, we have to calculate daily returns using a particular percentage formula
close = my_data['Close']
daily_returns = np.array([])

for i in range(1, 2518):
  R = 100*((close[i]/close[i - 1]) - 1)
  daily_returns = np.append(daily_returns, R)
  
Sneh = pd.DataFrame({'daily_returns': daily_returns})
print(Sneh)
```

```python
#ploting the histogram of daily returns
Sneh.plot(kind='hist')

#This looks like a normal continous distribution and support is 8. The distribution is skewed on the right side as the positive skewed. It does not have fat tails. For this continous probability distribution family would model this distribution
```

```python
#with the given data, by using fit method of scipy we will compute the Maximum Likelihood Estimators for the parameters of the distribution

import scipy
import scipy.stats as st

x = Sneh['daily_returns'].to_numpy()

#Here we have chosen quiet a wide variety of continous probability distribution family

list = ['logistic','hypsecant','norm','gamma','laplace', 'norminvgauss','skewnorm']
for name in list:
  dist = getattr(scipy.stats, name)
  param = dist.fit(x)
  x_sort = np.sort(x)
  pdf_fitted = dist.pdf(x_sort, *param[:-2], loc=param[-2], scale=param[-1])
  plt.figure()
  plt.hist(x, density=True, bins=100)
  plt.plot(x_sort,pdf_fitted)
  my_family_of_distribution = '%s, loc= %0.2f, scale=%0.2f' % (name, param[-2], param[-1])
  plt.title(my_family_of_distribution)

#Here we have plotted pdf along the histogram of daily return using the family of probabilty distribution. By result we have obtained from ploting the families of probability distribution, we can say that all of the functions are providing great result. Specially 'laplace', 'norminvgauss'
```

```python
# At last, we plot the QQ plot of the daily returns data versus the model we have just fitted above. Quality of the fit can be seen by the qq plots of most of the continous probability distribution family but specially for 'laplace', 'norminvgauss'. But from the qq plots we can confirm that from our chosen family 'norminvgauss' fits the best.

def standardize(daily_returns):
  z = (daily_returns-np.mean(daily_returns))/np.std(daily_returns)
  return z

def get_quantiles(data):
    quantiles = []
    for q in np.arange(0, 1.001, 0.001):
        quantiles.append(np.quantile(data, q))
    return quantiles    

def pyqqplot(x, param):
    plt.figure()
    x = standardize(x)
    param = standardize(param)
    q1 = np.array(get_quantiles(x))
    q2 = np.array(get_quantiles(param))
    plt.scatter(q1, q2 )

    minimum = min(x.min(), param.min())
    maximum = max(x.max(), param.max())
    plt.plot([minimum, maximum], [minimum, maximum], 'r-')
    plt.xlabel('daily returns')
    plt.ylabel('fitted model')
    plt.show()


list = ['logistic','hypsecant','norm','gamma','laplace', 'norminvgauss','skewnorm']
for name in list:
  dist = getattr(scipy.stats, name)
  param = dist.fit(x)
  x_sort = np.sort(x)
  pdf_fitted = dist.pdf(x_sort, *param)
  plt.figure()
  plt.hist(x, density=True, bins=100)
  plt.plot(x_sort,pdf_fitted)
  my_family_of_distribution = '%s, loc= %0.2f, scale=%0.2f' % (name, param[-2], param[-1])
  plt.title(my_family_of_distribution)
  y = dist.rvs(*param, size=10000)
  pyqqplot(x,y)
```
